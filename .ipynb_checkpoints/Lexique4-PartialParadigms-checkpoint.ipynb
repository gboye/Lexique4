### Correspondances des colonnes Lexique/Flexique

```python
finiMap={"ai":"pst.","pi":"prs.","fi":"fut.","ii":"ipfv.",
          "ps":"sbjv.","is":"pst.sbjv.",
          "pI":"imp.",
          "pc":"cond.",
         }
numMap={"S":"sg","P":"pl"}
persMap={p+n:p+"."+numMap[n] for p in "123" for n in "SP"}
genreNumMap={p+n:p.lower()+"."+numMap[n] for p in "MF" for n in "SP"}
pastpartMap={"pp"+gn:"pst.ptcp."+genreNumMap[gn] for gn in genreNumMap}
flexMap={t+p:finiMap[t]+persMap[p] for t in finiMap for p in persMap}
for k in ["pI1S","pI3S","pI3P"]:
    del flexMap[k]
flexMap.update({"inf":"inf","pP":"prs.ptcp"})
flexMap.update(pastpartMap)
flexMap
```

```python
def getFormes(row,freq=True):
    result=[]
    verbe=row["lexeme"]
    result.append(verbe)
#     print verbe
    for case in cases:
        if not freq or row[case]>0:
            if verbe in dfFlexique["lexeme"].values:
                vCase=dfFlexique.loc[dfFlexique.lexeme.isin([verbe]),flexMap[case]].values[0]
                if vCase!=u"#DEF#":
                    result.append(vCase)
                else:
                    result.append(np.NaN)
            else:
                result.append(np.NaN)
        else:
            result.append(np.NaN)
    return pd.Series(result)
```

### Encodage phonétique Flexique/Sampa

```python
nasalMap={u"ɛ̃":u"ê",u"ɑ̃":u"â",u"ɔ̃":u"ô"}
neutMap={u"ɛ":u"e",u"E":u"e",u"O":u"o",u"ɔ":u"o",u"ʁ":u"r",u"ə":u"@"}
def recoderIPA(chaine):
    result=chaine
    for k,v in nasalMap.iteritems():
        result=result.replace(k,v)
    for k,v in neutMap.iteritems():
        result=result.replace(k,v)
    return result

def recoderFormes(row):
    result=[]
    result.append(row["lexeme"])
    for case in cases:
        if row[case]==row[case]:
            result.append(recoderIPA(row[case]))
        else:
            result.append(row[case])
    return pd.Series(result)
    
```

```python
dfFormes=dfFormesBrutes.apply(recoderFormes,axis=1)
dfFormes.columns=["lexeme"]+cases
dfFormes.head()
```

### Réduction à un sous-paradigme
En réduisant à un sous-paradigme, on ne veut garder que des lexèmes qui ont au moins une forme dans la zone considérée :
- .dropna(thresh=2) permet de garantir qu'il y aura au moins deux valeurs dans la ligne
    - le lexème
    - une forme

```python
dfSubPar=dfFormes["lexeme inf ppMS pi3S ppFS pi1S pi3P".split()].dropna(thresh=2)
subCases=dfSubPar.columns.tolist()[1:]
print len(dfSubPar),len(dfFormes)
dfSubPar.head()
```

```python
from difflib import SequenceMatcher

def getMatch(string1,string2):
    return SequenceMatcher(None, string1, string2).find_longest_match(0, len(string1), 0, len(string2))

def getRadCF(row,lCases=subCases):
    result=[]
    result.append(row.lexeme)
    for case in lCases:
        if row[case]==row[case]:
            radical=row[case]
            break
#     radical=row[subCases[0]]
#     print radical
    for case in lCases[1:]:
#         print row[case],row[case]==row[case]
        if row[case]==row[case]:
            match=getMatch(radical,row[case])
            radical=radical[match.a: match.a + match.size]
    result.append(radical)
    for case in lCases:
        if row[case]==row[case] and radical!="":
            result.append(row[case].replace(radical,"-",1))
        elif row[case]==row[case] and radical=="":
            result.append(row[case])
        else:
            result.append(np.NaN)
    return pd.Series(result)
```

```python
dfSubCF=dfSubPar.apply(getRadCF,axis=1)
dfSubCF.columns=["lexeme","radical"]+subCases
```

```python
dfSubCF=dfSubCF.groupby(subCases).agg({"radical":"count","lexeme":list}).reset_index()
dfSubCF
```

```python
def getCF(row,lCases=subCases,lCF=dfSubCF):
    result=[]
    lexeme=row.lexeme
#     print lexeme,
    for ix,cf in lCF.iterrows():
#         print ix
        bCF=True
        bRad=False
        for case in lCases:
            formeLex=row[case]
            if formeLex==formeLex and cf[case]==cf[case]:
#                 print cf[case]
                formeCF=cf[case].replace(u"-",u"(.*)")+u"$"
                m=re.search(formeCF,formeLex)
                if not m:
                    bCF=False
#                     print "allomorphie affixale"
                    break
                elif not bRad:
                    if m.groups():
                        rad=m.group(1)
                    else:
                        rad=""
#                     print rad,
                    bRad=True
                elif not m.groups() or m.group(1)!=rad:
                    bCF=False
#                     print "allomorphie radicale"
                    break                    
        if bCF:
            result.append(unicode(ix))
#             print ix,
#     print
    return pd.Series([lexeme,",".join(result),len(result),result])
```

```python
dfSubParAmbiguities=dfSubPar[pd.isnull(dfSubPar).any(axis=1)].apply(getCF,axis=1)
dfSubParAmbiguities.columns=["lexeme","CF","nCF","lCF"]
```

```python
dfTotPar=dfFormes[["lexeme"]+cases48].dropna(thresh=2)
dfTotPar
```

```python
dfTotPatrons=dfTotPar.apply(lambda x: getRadCF(x,cases48),axis=1)
dfTotPatrons.columns=["lexeme","radical"]+cases48
dfTotPatrons
```

```python
dfTotCF=dfTotPatrons.groupby(cases48).agg({"radical":"count","lexeme":list}).reset_index()
dfTotCF
```

```python
def getIncrementalCF(thresh,lDF=dfTotPatrons,lCases=cases48,lCF=dfTotCF):
    lDF1=lDF.dropna(thresh=thresh)
#     display(df1)
    lDF2=lDF.dropna(thresh=thresh-1).drop(lDF1.index)
    if len(lDF2)>0:
        lDF3=lDF2.apply(lambda x: getCF(x,lCases=lCases,lCF=lCF),axis=1)
        lDF3.columns=["lexeme","CF","nCF","lCF"]
    else:
        lDF3=pd.DataFrame(columns=["lexeme","CF","nCF","lCF"])
    return lDF3
```

```python
dfTotAmbiguities=pd.DataFrame(columns=["lexeme","CF","nCF","lCF"])
dfTotAmbiguities
```

```python
for i in range(49):
    print i,
    dfInc=getIncrementalCF(50-i,lCF=dfTotCF)
#     print "dfInc"
#     display(dfInc)
    dfNewCF=dfTotPatrons.loc[dfTotPatrons.index.isin(dfInc[dfInc.nCF==0].index)].fillna("XXX").groupby(cases48).agg({"radical":"count","lexeme":list}).reset_index().replace(u"XXX",np.NaN)
    display(dfNewCF)
    dfTotCF=pd.concat([dfTotCF,dfNewCF],ignore_index=True)
#     print "dfTotCF"
#     display(dfTotCF)
    dfLocAmbiguities=dfInc[dfInc.nCF>0]
#     print "dfLocAmbiguities"
#     display(dfLocAmbiguities)
    dfTotAmbiguities=pd.concat([dfTotAmbiguities,dfLocAmbiguities],ignore_index=True)
#     print "dfTotAmbiguities"
#     display(dfTotAmbiguities)
display(dfTotCF[["lexeme","radical"]+cases48])
display(dfTotAmbiguities)
```

```python
def multi2simple(lDF):
    lDF.columns=[c if c else sc for sc,c in lDF.columns.to_series().tolist()]
    lDF.reset_index(inplace=True)
    return lDF
```

# Faire des cliques de CF compatibles
- associer le nombre de paires identiques comme score à chaque lien

```python
import networkx as nx
import itertools as it
compatibleCF=nx.Graph()
for combi in it.combinations(dfTotCF.index,2):
#     display(fusionCF(dfTotCF[dfTotCF.index.isin(combi)]))
    if len(fusionCF(dfTotCF[dfTotCF.index.isin(combi)]))>0:
        compatibleCF.add_edge(combi[0],combi[1])
```

Parmi les classes flexionnelles identifiées, quelques unes sont unifiables :
- 10 et 12 : devenir, venir
- 14, 15 et 19 : rester, laisser, donner
- 28 et 29
- 40 et 41
- 45 et 48
- 54 et 56

```python
def fusionCF(lDF,lCases=cases48):
#     print len(lDF)
    bCompatible=True
    fusedCF={}
    for ix,row in lDF[lCases].iterrows():
        for case in lCases:
            if case not in fusedCF:
                fusedCF[case]=row[case]
            elif row[case]==row[case]:
                if fusedCF[case]==fusedCF[case]:
                    if row[case]!=fusedCF[case]:
                        bCompatible=False
#                         print ix,case,row[case],fusedCF[case]
                        fusedCF=None
                        break
                else:
                    fusedCF[case]=row[case]
        if not bCompatible:
            break
    return pd.Series(fusedCF)
```
