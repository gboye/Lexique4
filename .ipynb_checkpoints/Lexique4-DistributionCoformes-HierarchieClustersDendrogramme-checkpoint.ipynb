### Faire l'histogramme du ratio CMB caseLex/caseVoisins
- dfAX la distribution de la fréquence dans les cases connues pour le premier lexème
- dfFX la distribution de la fréquence dans les cases connues pour les autres lexèmes

On remplit les NaN avec des 0 pour calculer la distribution moyenne pour chaque case pour les autres lexèmes.
- dfFX**.fillna(0)**.mean(axis=1)

On remplace les moyennes à 0 par des NaN pour éviter les divisions par 0 et signaler les points sans données pour l'histogramme.
- dfFX.fillna(0).mean(axis=1)**.replace(0,np.NaN)**

On divise la distribution de fréquence du premier lexème par la moyenne des autres et on remplace les NaN par des nombres négatifs pour distinguer les 0 qui sont liés à la présence effective de données des NaN qui indiquent l'absence d'information.
- (dfAX/dfFX.fillna(0).mean(axis=1).replace(0,np.NaN))**.replace(np.NaN,-.5)**

Les barres au dessus du ratio ratioLim sont colorées en vert, celles en dessous du ration en orange, les points sans données sont signalés par des barres rouges

```python
def plotFreqRatioCMB(famille,gParadigme,ratioLim=0.2):
    dfAX=(gParadigme[gParadigme["lexeme"].isin([famille[0]])].set_index("lexeme").loc[famille[0]][etCols].T)
    dfFX=(gParadigme[gParadigme["lexeme"].isin(famille[1:])].set_index("lexeme").loc[famille[1:]][etCols].T)
    dfAX=(dfAX/dfFX.fillna(0).mean(axis=1).replace(0,np.NaN)).replace(np.NaN,-1)
#     display(dfAX)
    dfAX.name="No Data"
    ax=dfAX.plot(kind="bar",
                 figsize=(40,5),
                 color=np.where(dfAX>ratioLim, 'g', (np.where(dfAX<0, 'r','orange'))),
                )
    colorerXTicks(ax)
    ax.set(ylim=(-1,None),ylabel=u"ratio lexème/voisin")
    ax.legend(loc="upper left")
    leg = ax.get_legend()
    leg.legendHandles[0].set_color('red')

    jointFamille="".join([l.capitalize() for l in famille])
#     plt.savefig(repHDR+u'Lex4-BP-%s-VerbeExpFreq.pdf'%(jointFamille), dpi=300, bbox_inches="tight")
    plt.show()
```

```python
def preparerDataRaw(gParadigme):
    paradigme=gParadigme[etCols].copy()
    if "lexeme" in paradigme.columns:
        lexemeLabels=paradigme["lexeme"].tolist()
        del paradigme["lexeme"]
    if "freqLexeme" in paradigme.columns:
        del paradigme["freqLexeme"]

    if "case" in paradigme.columns:
        del paradigme["case"]
    paradigme.fillna(value=0,inplace=True)

    data_raw=paradigme.iloc[:].values
    return data_raw
```

### Clusterisation
- getLinkage permet de calculer la structure de clusterisation par proximité entre les différents vecteurs
    - les disctributions d'occurrences les plus proches sont clusterisées ensemble
        - soit d'après la répartition normalisée (normalize=True)
        - soit d'après les valeurs brutes (normalize=False)
- scipy.cluster.hierarchy.linkage fait le calcul de la structure arborescente en clusters
    - "ward" est la méthode conseillée par défaut

```python
from scipy.cluster import hierarchy as hc
```

```python
def getLinkage(gParadigme,normalize=True):
    data_raw=preparerDataRaw(gParadigme)
    if normalize:
        data=normaliserDataRaw(data_raw)
    else:
        data=data_raw
    Z=hc.linkage(data,method='ward')
    return Z
```

## Exploitation de la hiérarchie complète
On peut utiliser networkx pour représenter la hiérarchie dichotomique Z obtenue avec [hc.linkage](https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html#scipy.cluster.hierarchy.linkage) :
- [hc.to_tree(Z)](https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.to_tree.html) donne un arbre qui contient la hiérarchie sous la forme d'un [ClusterNode](https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.ClusterNode.html#scipy.cluster.hierarchy.ClusterNode), une structure qui contient pour chaque nœud :
    - id, le label/nom du nœud
    - left, le ClusterNode du descendant à gauche
    - right, le ClusterNode du descendant à droite
- écriture d'un graphe orienté correspondant à la hiérarchie
    - addLevel(root)
        - procédure récursive qui copie la structure de l'arbre dans un graphe orienté *clusterHierarchy*
- extraction de l'ancêtre commun à une paire de verbes :
    - nx.algorithms.lowest_common_ancestor(G,Verbe1,Verbe2)
- extraction de la liste des verbes descendant d'un ancêtre commun
    - nx.descendants(G,ancestor), liste de tous les descendants (y compris les nœuds intermédiaires)
    - G.out_degree(n)==0, sélectionne les nœuds qui n'ont pas de descendants (les feuilles)

```python
clusterHierarchy=nx.DiGraph()
```

```python
def addLevel(G,currentNode):
    if currentNode.left:
        addLevel(G,currentNode.left)
        G.add_edge(currentNode.id,currentNode.left.id)
    if currentNode.right:
        addLevel(G,currentNode.right)
        G.add_edge(currentNode.id,currentNode.right.id)
    return
```

```python
root=hc.to_tree(Z)
addLevel(clusterHierarchy,root)
nbNiveauxHierarchy=nx.dag_longest_path_length(clusterHierarchy)
print u"La hiérarchie a %d niveaux."%nbNiveauxHierarchy
```

### Trouver les voisinages

#### Ancêtre et parents

```python
def getAncestor(node,G=clusterHierarchy):
    result=[a for a,b in G.edges() if b==node]
    if len(result)==1:
        return result[0]
    elif len(result)==0:
        print "no ancestor for",node
        return node
    else:
        print "more than one ancestor for",node
        return None

def getDescendants(node,G=clusterHierarchy):
    descendantNodes=nx.descendants(G,node)
    return [n for n in descendantNodes if G.out_degree(n)==0]

def getCluster(node,level=1,G=clusterHierarchy):
    ancestor=node
    for l in range(level):
        ancestor=getAncestor(ancestor,G)
    return getDescendants(ancestor,G)
```

#### Extraction du premier ancêtre commun
On peut utiliser networkx pour extraire le premier ancêtre commun.
- écriture d'un graphe orienté correspondant à la hiérarchie
    - add_edge(père,fils)
- extraction de l'ancêtre
    - nx.algorithms.lowest_common_ancestor(G,Verbe1,Verbe2)

```python
def getCommonAncestor(v1,v2,G=clusterHierarchy):
    return nx.algorithms.lowest_common_ancestors.lowest_common_ancestor(G,v1,v2)
```

```python
def getCommonClusterFamily(v1,v2,G=clusterHierarchy):
    ancestor=getCommonAncestor(v1,v2,G)
    descendantNodes=nx.descendants(G,ancestor)
    return [n for n in descendantNodes if G.out_degree(n)==0]
```

# Test de l'hypothèse de Copot, Mickus, Bonami
le ratio entre la fréquence relative d'une case pour un lexème et la fréquence relative moyenne pour ses voisins donne une indication de la défectivité de cette case pour ce lexème.

En dessous d'une certaine valeur du ratio ,par exemple 0.1, en moyenne les voisins ont une forme 10 fois plus fréquente que celle observée pour le lexème observé. 

On pourrait aussi utiliser ce calcul pour des ratios très très faibles ou 0.

Le problème est l'existence des données pour les voisins. Si il n'y a pas de voisin absolument fréquent, la plupart des cases du paradigme ne seront pas du tout remplies pour les voisins non plus et on ne pourra rien conclure...

Dans le cas ci-dessous, on observe les voisins sémantiques de clore d'après un dictionnaire de synonymes en prenant 5 voisins. ## Clusterisation par KMeans

### Choix du nombre de clusters
Calcul du WCSS (within cluster sum of squares)
kmeans.inertia_ est la valeur du wcss du modèle.

```python
from sklearn.cluster import KMeans
wcss={}
for i in range(1,55):
    kmeans=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10)
    kmeans.fit(data)
    wcss[i]=kmeans.inertia_
```

```python
dfWCSS=pd.DataFrame.from_dict(wcss,orient="index")
dfWCSS.plot(figsize=(10,5))
```

```python
dfWCSS.diff().plot(figsize=(20,5),xticks=(np.arange(0, 50, step=1))
)
```

### Clusterisation autour de K points
k-means++ est une stratégie d'initialisation qui permet d'éviter une trappe d'optimisation locale

```python
nbClusters=7
kmeans=KMeans(n_clusters=nbClusters,init="k-means++")
y_kmeans=kmeans.fit_predict(data)
```

```python
lexemeType={}
for nY,y in enumerate(y_kmeans):
    lexemeType[lexemeLabels[nY]]=y
specialLexemes=boyeDefectifs#[u"être",u"avoir",u"pouvoir",u"savoir",u"faire"]
for lex in specialLexemes:
    if lex in lexemeType:
        print lex,lexemeType[lex]
```

```python
allLexemes=True
typeLexemes = {}
for k, v in lexemeType.iteritems():
    if k in specialLexemes or allLexemes:
        typeLexemes[v] = typeLexemes.get(v, [])
        typeLexemes[v].append(k)
```

### plotter les distributions correspondant aux K centres

```python
pdKMeans=pd.DataFrame(kmeans.cluster_centers_,columns=cases).transpose()
```

```python
for i in range(nbClusters):
    iSum=float(pdKMeans[i].sum())
#    print iSum
    pdKMeans[i]=pdKMeans[i]/iSum
#    print pdKMeans[i]
    ax=pdKMeans[i].plot(kind="bar",figsize=(10, 3))
    ax.set_ylim([0,1])
    specLexI=[]
    for l in specialLexemes:
        if l in typeLexemes[i]:
            specLexI.append(l)
            if len(specLexI)>=10:
                break
    nPadLex=5-len(specLexI)
    if nPadLex>0:
        specLexI=specLexI+typeLexemes[i][:nPadLex]
    plt.title("Type %d => %d: "%(i,len(typeLexemes[i]))+",".join(specLexI)+u"\n")
    plt.show()
```
